## Outliers

We check outliers on NumberOfItemsPurchased and CostPerItem

```python
# Create Outliers Check Function

def outliers_check(df, column):
    # Calculate Q1 and Q3
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    # lower and upper bound
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    # show outliers
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers
```

## Normality Test

### Shapiro-Wilk Test

```python
# Shapiro-Wilk Test for NumberOfItemsPurchased
from scipy.stats import shapiro
stat, p = shapiro(df["NumberOfItemsPurchased"])
print("stat=%.3f, p=%.3f" % (stat, p))
if p > 0.05:
    print("Fail to reject H0 : NumberOfItemsPurchased is normally distributed")
else:
    print("H0 rejected : NumberOfItemsPurchased is not normally distributed")
```

```python
# Shapiro-Wilk Test for CostPerItem
from scipy.stats import shapiro
stat, p = shapiro(df["CostPerItem"])
print("stat=%.3f, p=%.3f" % (stat, p))
if p > 0.05:
    print("Fail to reject H0 : CostPerItem is normally distributed")
else:
    print("H0 rejected : CostPerItem is not normally distributed")
```

### KS Test

```python
# KS Test
from scipy.stats import kstest
stat, p = kstest(df["NumberOfItemsPurchased"], 'norm')
print("stat=%.3f, p=%.3f" % (stat, p))
if p > 0.05:
    print("Fail to reject H0 : NumberOfItemsPurchased is normally distributed")
else:
    print("H0 rejected : NumberOfItemsPurchased is not normally distributed")
```

```python
# KS Test
from scipy.stats import kstest
stat, p = kstest(df["CostPerItem"], 'norm')
print("stat=%.3f, p=%.3f" % (stat, p))
if p > 0.05:
    print("Fail to reject H0 : CostPerItem is normally distributed")
else:
    print("H0 rejected : CostPerItem is not normally distributed")
```

## 7. RFM Score

```python
# Show RFM
display(rfm)
```

```python
# Getting RFM statistics
rfm.describe()
```

```python
import pandas as pd

# Function to count RFM score
def rfm_score(baris):
    # Recency
    if baris['Recency'] <= 17:
        r = 4
    elif baris['Recency'] <= 50:
        r = 3
    elif baris['Recency'] <= 141:
        r = 2
    else:
        r = 1

    # Frequency
    if baris['Frequency'] <= 1:
        f = 1
    elif baris['Frequency'] <= 2:
        f = 2
    elif baris['Frequency'] <= 5:
        f = 3
    elif baris['Frequency'] <= 10:
        f = 4
    else:
        f = 5

    # Monetary
    if baris['Monetary'] <= 1234:
        m = 1
    elif baris['Monetary'] <= 2670:
        m = 2
    elif baris['Monetary'] <= 6479:
        m = 3
    elif baris['Monetary'] <= 10000:
        m = 4
    else:
        m = 5

    return r, f, m

# Add column R_Score, F_Score, M_Score
rfm[['R_Score', 'F_Score', 'M_Score']] = rfm.apply(lambda baris: pd.Series(rfm_score(baris)), axis=1)

# Add column RFM_Score
rfm['RFM_Score'] = rfm['R_Score'] + rfm['F_Score'] + rfm['M_Score']

# Showing RFM
rfm
```

```python
# RFM Score Histogram

sns.histplot(rfm['RFM_Score'])
plt.show()
```

## 8. RFM Segmentation

```python

rfm['FM_Score'] = rfm['F_Score'] + rfm['M_Score']

# Manual Clustering using IF-ELSE Conditions
def segment_customer(row):
    recency = row['R_Score']
    fm_combined = row['FM_Score']
    if recency >= 4 and fm_combined >= 4:
        return 'Champions'
    elif recency >= 3 and fm_combined >= 3:
        return 'Loyal Customers'
    elif recency >= 2 and fm_combined >= 3:
        return 'Potential Loyalist'
    elif recency >= 3 and fm_combined <= 2:
        return 'Promising'
    elif recency < 2 and fm_combined >= 3:
        return 'At Risk'
    elif recency < 2 and fm_combined <= 2:
        return 'Lost'
    else:
        return 'Uncategorized'

# Apply the segmentation function
rfm['Customer_Segment'] = rfm.apply(segment_customer, axis=1)
rfm
```

```python
df["Acquisition Date"]= df.groupby("UserId")["TransactionTime"].transform('min').dt.to_period('M')
df
```

## 2. Transaction Month

```python
df['Transaction Month']= df["TransactionTime"].dt.to_period('M')
df
```

## 3. New User Every Month

```python
cohort_table = df.groupby(['Acquisition Date','Transaction Month'])['UserId'].nunique().reset_index()
cohort_table = cohort_table.rename(columns = {'UserId':'N_Users'})
cohort_table
```

## 4. Period Month

```python
# Counting Months
cohort_table['Period Month'] = (
    (cohort_table['Transaction Month'].dt.year - cohort_table['Acquisition Date'].dt.year) * 12 +
    (cohort_table['Transaction Month'].dt.month - cohort_table['Acquisition Date'].dt.month)
)

cohort_table
```

## 5. Pivot Table

```python
cohort_pivot = cohort_table.pivot_table(index='Acquisition Date', columns='Period Month', values='N_Users')
cohort_pivot
```

## 6. Set as Percentage

```python
cohort_pivot_per = cohort_pivot.apply(lambda x: x/x.iloc[0]*100, axis = 1)
cohort_pivot_per
```
